```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY")
```

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
documents = SimpleDirectoryReader("data").load_data()
```

```python
documents
```

```python
index = VectorStoreIndex.from_documents(documents,show_progress=True)
```

```python
index
```

```python
query_engine = index.as_query_engine()
query_engine
```

```python
response = query_engine.query("Who is the author")
response
```

```python
response2 = query_engine.query("What is the name of the autyhor")
print(response2)
```

```python
from llama_index.core.response.pprint_utils import pprint_response
pprint_response(response2,show_source=True)
```

```python
response3 = query_engine.query("what is attention is all you need")
print(response3)
```

```python
pprint_response(response3,show_source=True)
```

```python
response4 = query_engine.query("Who are the publishers of attention is all you need")
print(response4)
```

```python
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.indices.postprocessor import SimilarityPostprocessor
```

```python
retriever = VectorIndexRetriever(index=index, similarity_top_k=4)
postProcessor = SimilarityPostprocessor(similarity_cutoff=0.80)
query_engine = RetrieverQueryEngine(retriever=retriever, node_postprocessors=[postProcessor])
```

```python
response_new = query_engine.query("What is attention is all you need")
pprint_response(response_new, show_source=True)
```

```python
import os.path
from llama_index.core import StorageContext, load_index_from_storage
```

```python
PERSIST_DIR = "./storage"

if not os.path.exists(PERSIST_DIR):
    documents = SimpleDirectoryReader("data").load_data()
    index = VectorStoreIndex(documents)
    index.storage_context.persist(persist_dir=PERSIST_DIR)
else:
    storage_context = StorageContext.from_defaults(persist_dir=PERSIST_DIR)
    index = load_index_from_storage(storage_context=storage_context)
```

```python
query_engine = index.as_query_engine()
```

```python
print(query_engine.query("what are transformers"))
```

```python
print(query_engine.query("where was the call center located"))
```

```python
pprint_response(query_engine.query("What is the plot"),show_source=True)
```

```python
pprint_response(query_engine.query("what is brain tumor"),show_source=True)
```
